1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_func(x):
    return 10*x-5
```

```python
def df(f, h=0.0001):
  return (f(x + h) - f(x)) / h
```

```python
x = sp.symbols('x') # Визначаємо символічну змінну, щоб продемонструвати саме формулу похідної
df(my_func)
```

```python
df(my_func,5) #на прикладі
```

```python
sp.diff(my_func(x), x) #або ж одразу визначаємо похідну за допомогою бібліотеки
```

```python
x = np.linspace(0,10,20)

plt.figure(figsize=(10, 5))
plt.plot(x, my_func(x), label='f(x)', color='blue')
plt.plot(x, df(my_func,x), label="f'(x)", color='orange')
plt.legend()
```

Висновок: З кожною вивченою годиною оцінка збільшується на 10 (одна година рівна 10 балам)

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import numpy as np
import matplotlib.pyplot as plt

x_sp = sp.symbols('x_sp')
f_sp = sp.sin(x_sp)
df_sp= sp.diff(f_sp ,x_sp)
print("Функція:", f_sp)
print("Похідна:", df_sp)

# Перетворення у числові функції
f = sp.lambdify(x_sp, f_sp, modules=['numpy'])
df = sp.lambdify(x_sp, df_sp, modules=['numpy'])

x = np.linspace(0, 5, 100)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f(x), label=f'f(x) = {f_sp}', color='blue')
plt.plot(x, df(x), label=f"f'(x) = {df_sp}", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_sp = sp.symbols('x_sp')
f_sp = sp.ln(x_sp)
df_sp= sp.diff(f_sp ,x_sp)
print("Функція:", f_sp)
print("Похідна:", df_sp)

# Перетворення у числові функції
f = sp.lambdify(x_sp, f_sp, modules=['numpy'])
df = sp.lambdify(x_sp, df_sp, modules=['numpy'])

x = np.linspace(0, 5, 100)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f(x), label=f'f(x) = {f_sp}', color='blue')
plt.plot(x, df(x), label=f"f'(x) = {df_sp}", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
import numpy as np
import matplotlib.pyplot as plt

x_sp = sp.symbols('x_sp')
f_sp = x_sp**2 - 5*x_sp + 1
df_sp= sp.diff(f_sp ,x_sp)
print("Функція:", f_sp)
print("Похідна:", df_sp)

# Перетворення у числові функції
f = sp.lambdify(x_sp, f_sp, modules=['numpy'])
df = sp.lambdify(x_sp, df_sp, modules=['numpy'])

x = np.linspace(0, 100, 100)

# Створення графіку
plt.figure(figsize=(20, 8))
plt.plot(x, f(x), label=f'f(x) = {f_sp}', color='blue')
plt.plot(x, df(x), label=f"f'(x) = {df_sp}", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
